```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sigma = 10
gamma = 1/np.sqrt(1-.3**2)
theta = sigma/200
```

```python
def H(u):
    return(-u+np.log(1.+u+u**2/2))
def G(g):
    return( 1-np.e**(H(g/theta) - H(1/theta)) )
def X(u):
    return(np.log(1-u) + H(1/theta))
```

```python
base = 10.
powers = np.arange(-16,0)
samples = 10**2
u1 = []
for p in powers:
    u1 = np.append(u1,np.arange(1-base**p,1-base**(p+1),((1-base**(p+1))-(1-base**(p)))/samples))
x = X(u1)
plt.plot(u1,x)
plt.gcf().set_facecolor('white')
plt.gcf().set_size_inches(19,18)
xLo=x[0]
xHi=x[-1]
print("Lowest X =", xLo)
print("Highest X =",xHi)
```

```python
samples = 10**3
x = np.arange(0,60,60/samples)
h = H(x)
plt.plot(h,x)
plt.gcf().set_facecolor('white')
plt.gcf().set_size_inches(18,16)
```

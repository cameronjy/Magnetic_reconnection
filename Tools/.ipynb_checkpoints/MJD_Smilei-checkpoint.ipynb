```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sigma = 10 #Magnetization
theta = sigma/200 #Theta = kT/mc^2. Temperature Parameter.
```

```python
def analyticMJD(g): #Analytic Maxwell Juttner Distribution. f(gamma) = gamma^2 * beta * e^(-gamma/theta). beta = sqrt(1-gamma^-2).
    return(g**2*np.sqrt((1-g**-2))*np.e**(-g/theta))
def G(g): #Cumulative distribution of proposed distribution g(x). Written in terms of H.
    return( 1-np.e**(H(g/theta) - H(1/theta)) )
def H(u): #Function that G is written in terms of. G^-1 = theta * H^-1(X(u)), where function X(u) is the next definition in this file.
    return(-u+np.log(1.+u+u**2/2))
def X(u): #Function of u in [0,1) that gets plugged in to H^-1 to produce G^-1.
    return(np.log(1-u) + H(1/theta))
```

```python
#The functions in this cell depend on these precomputed values.
xLow = min((X(0),X(1-10**-16))) #Lowest value of function X.
gTable = np.arange(0,-xLow*1.5,(0-(xLow*1.5))/10**3) #Table of possible gamma values.
hTable = H(gTable) #Table of possible H values.
hMax = max(hTable)
hMin = min(hTable)
#Linear interpolation using the values of gTable and hTable. Function computes H^-1(Y) for a single number or a set of numbers.
def Hi(Y):
    if(type(Y) == np.ndarray):
        z = np.zeros(Y.size)
        for i in range(Y.size): #For each given Y
            if( Y[i] <= hMin or Y[i] >= hMax ): #Quit if any y in Y is out of bounds.
                print("Out of Bounds")
                return();
            for j in range(hTable.size):
                if(hTable[j] < Y[i]): #Look through the hTable to find the first value that is less than a given y in Y.
                    z[i] = gTable[j-1] + ((gTable[j]-gTable[j-1])/(hTable[j]-hTable[j-1]))*(Y[i]-hTable[j-1]) #Linearly interpolate ahead and stop looking for h < y.
                    break
    else:
        if( Y <= hMin or Y >= hMax ):
            print("Out of Bounds")
            return();
        for j in range(hTable.size):
            if(hTable[j] < Y):
                z = gTable[j-1] + ((gTable[j]-gTable[j-1])/(hTable[j]-hTable[j-1]))*(Y-hTable[j-1])
                break
    return (z)
#Acceptance rejection scheme. Produces n values of gamma distributed along the MJ distribution. The function repeats following setps n times:
#1. Computes gamma = G^-1(u1) where for a random u1 in [0,1)
#2. Selects a random u2 in [0,1)
#3. Accepts gamma if U2 < sqrt(1 - gamma^-2) = beta = f(gamma)/g(gamma) by adding the value to the list of n gammas, returns to step 1 otherwise.
def MJD(n):
    g = np.zeros(n)
    for i in range(g.size):
        u2 = 1
        p = 0
        while u2 > p: #Check the condition of step 3.
            u1 = np.random.rand()
            gamma1 = theta*Hi(X(u1)) #Step 1.
            u2 = np.random.rand() #Step 2.
            if(gamma1 >= 1): #Occaisonally Hi gives a value less than 1.
                p = np.sqrt(1-gamma1**-2) #Step 3.
            else:
                p = 0
                u2 = 1
        g[i] = gamma1
    return(g)
```

```python
points = 10**6
gammas = MJD(points) #Accepted gammas
cmjd,gamma1 = np.histogram(gammas,bins=200,density=True) #Bins values of accepted gammas into a distribution whose integral is equal to 1. cmjd = computed MJ distribution.
for i in range(gamma1.size-1): #This loop computes the midpoints of each histogram bin.
    gamma1[i] =+(gamma1[i+1]+gamma1[i])/2
    i += 1
gamma1 = gamma1[0:gamma1.size-1]
gamma2 = np.arange(1,max(gamma1),(max(gamma1)-1)/200) #Domain of distribution is determined by the largest gamma produced by MJD().
amjd = analyticMJD(gamma2) #Analytic MJ distribution.
amjd = amjd/(sum(amjd)*(max(gamma1)-1)/200) #Noramlize amjd integral to be equal to 1.

plt.rcParams['figure.figsize'] = [16,9]
plt.plot(gamma2,amjd,color='royalblue',linestyle='dashed')
plt.plot(gamma1,cmjd,'darkred')
plt.xlim(left=.9)
plt.ylim(0,1.1*max(cmjd))
plt.rcParams.update({'font.size':18})
plt.title('Acceptance-Rejection Maxwell-Juttner Distribution')
plt.legend(('Analytic','Acceptance-Rejection'))
plt.gca().set_xlabel('Gamma');
plt.gca().set_ylabel('f');
plt.gcf().set_facecolor('white')
```
